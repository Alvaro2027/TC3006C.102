teorema del Limite central TLC

dada una muestra de datos $X_1,X_2,...X_n$ de media $_mu$ y varianza $\sigma$ (no importa la distribucion), si$n$ es sificientemente grande entonces $\overline{X}~N(\MU,\frac{\sigma^2}{n})$


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lambda_parametro=0.5
t_muestra=10000 #numero de variables por muestra
num_muestras=1000 #numero de muestras
muestras=np.random.exponential(scale=(1/lambda_parametro), size=(num_muestras, t_muestra))
```

```python
muestras
```

```python
#calcular el promedio por muestra
medias_muestrales=np.mean(muestras,axis=1)
medias_muestrales
```

```python
#crear dataframe para la media de las muestras
data={'Muestra':range(1,num_muestras+1),'Media muestral':medias_muestrales}
df=pd.DataFrame(data)
df
```

```python
#crear histograma con las medias muestrales
plt.hist(medias_muestrales,bins=30,alpha=0.6,color='b',label='Medias Muestrales')
plt.axvline(x=1/lambda_parametro,color='r',linestyle='dashed',linewidth=2,label='mu')
plt.xlabel('valores de medias muestrales')
plt.ylabel('frecuencia de las medias muestrales (probabilidad)')
plt.title('Teorema de limite central TLC')
plt.legend()
plt.show()
```

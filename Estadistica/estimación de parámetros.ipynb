```python
df=pd.read_csv('/content/baterias(1).csv')
```

```python
df.columns
```

```python
import numpy as np
#imporatmos exponencial y normal para compara

from scipy.stats import norm, expon

# Generamos rangos de valores para las distribuciones teóricas
x = np.linspace(min(df['tiempo_hasta_agotar']), max(df['tiempo_hasta_agotar']), 100)

# Ahora calculamos parámetros para las distribuciones teóricas
mean_normal = df['tiempo_hasta_agotar'].mean()
std_normal = df['tiempo_hasta_agotar'].std()
lambda_exponential = 1 / df['tiempo_hasta_agotar'].mean()

# Por ultimo graficamos los histogramas y distribuciones teóricas y las comparamos (la exp y la normal)
plt.hist(df['tiempo_hasta_agotar'], bins=20, edgecolor='black', alpha=0.7, density=True, label='Data')
plt.plot(x, norm.pdf(x, mean_normal, std_normal), label='Normal Distribution')
plt.plot(x, expon.pdf(x, scale=1/lambda_exponential), label='Exponential Distribution')

plt.xlabel('tiempo_hasta_agotar')
plt.ylabel('Densidad')
plt.title('Histograma y Distribuciones Teóricas')
plt.legend()
plt.show()
```

Aqui vemos que la distribucion exponencial describe mejor los datos.

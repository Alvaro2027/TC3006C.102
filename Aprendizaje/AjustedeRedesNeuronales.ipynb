## 1. Evalúa con validación cruzada un modelo pereceptrón multicapa para las variables que se te asignaron para este ejercicio.

Utilizamos esta opcion

Modelo de regresión con red neuronal perceptrón multicapa:
porque la variable que queremos predecir  es continua y se busca predecir un valor numérico


```python
from sklearn import datasets
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
y = df.iloc[:, 0]
x = df.iloc[:, 1:]
n_features = x.shape[1]
print(n_features)

targets = y.unique()
n_clases = len(targets)
n_clases
```

```python
# Fit regression model using the complete dataset and calculate MSE and MAE
regr = MLPRegressor(hidden_layer_sizes=(20, 20), max_iter=20000)
regr.fit(x, y)

y_pred = regr.predict(x)
print('MSE: ', mean_squared_error(y, y_pred))
print("MAE: ", mean_absolute_error(y, y_pred))

# Evaluate regression model using k-fold cross-validation
n_folds = 3
# Set a random seed for reproducibility
np.random.seed(42)
kf = KFold(n_splits=n_folds, shuffle = True,  random_state=42)

mse_cv = []
mae_cv = []
for train_index, test_index in kf.split(x):

    # Training phase
    x_train = x.iloc[train_index]
    y_train = y.iloc[train_index]

    regr_cv = MLPRegressor(hidden_layer_sizes=(20, 20), max_iter=20000)
    regr_cv.fit(x_train, y_train)

    # Test phase
    x_test = x.iloc[test_index]
    y_test = y.iloc[test_index]

    y_pred = regr_cv.predict(x_test)

    # Calculate MSE and MAE

    mse_i = mean_squared_error(y_test, y_pred)
    print('mse = ', mse_i)
    mse_cv.append(mse_i)

    mae_i = mean_absolute_error(y_test, y_pred)
    print('mae = ', mae_i)
    mae_cv.append(mae_i)



# Calculate and print the average MSE and MAE across all folds
avg_mse = np.mean(mse_cv)
print(f"Average MSE across all folds: {avg_mse}")
avg_msa = np.mean(mae_cv)
print(f"Average MAE across all folds: {avg_msa}")

```

Un mayor valor promedio de MSE (112412.70) y MAE (191.97) en la validación cruzada k-fold en comparación con el MSE inicial (13180.30) y el MAE (88.93) sugieren que mi modelo podría no generalizar bien a datos no vistos. Esto significa que el rendimiento del modelo en los datos de entrenamiento parece ser mejor que su rendimiento en nuevos datos, lo que sugiere un posible problema de sobreajuste.

En otras palabras, los resultados de la validación cruzada indican que mi modelo no se desempeña tan bien en datos que no ha visto previamente como lo hace en los datos en los que fue entrenado. Esto podría deberse a que el modelo está demasiado ajustado a los datos de entrenamiento y no puede generalizar de manera efectiva a nuevos datos. Para mejorar el rendimiento de mi modelo en datos no vistos, podría considerar ajustar los hiperparámetros, aumentar el tamaño del conjunto de datos si es posible o explorar diferentes algoritmos de regresión.

## 2. Agrega al conjunto de datos columnas que representen los cuadrados de las variables predictoras (por ejemplo, M2, W2), así como los productos entre pares de variables (por ejemplo, PxS, MxW). Evalúa un modelo perceptrón multicapa para este nuevo conjunto de datos.

```python
import pandas as pd
import numpy as np

# Definir las variables originales
x_variables = ['M', 'W', 'H', 'P', 'S']

# Extraer las variables predictoras y la variable de respuesta
X = df[x_variables].values
y = df['VR'].values

# Agregar las variables al cuadrado
X_squared = X ** 2

# Extender la lista de nombres x_variables con los nombres de las variables al cuadrado
x_squared_variable_names = [variable + '_squared' for variable in x_variables]
x_variables.extend(x_squared_variable_names)

# Combinar X y X_squared
X = np.column_stack((X, X_squared))

# Comparar 'M' con 'M_squared'
comparison_df = pd.DataFrame({'M': X[:, x_variables.index('M')], 'M_squared': X[:, x_variables.index('M_squared')]})

print(comparison_df.head())
```

```python
# Agregar interacciones solo entre las variables originales
new_variable_names = []

num_predictors = X.shape[1]
num_original_variables = len(x_variables)

for i in range(num_original_variables):
    for j in range(i + 1, num_original_variables):
        # Verificar que ambas variables involucradas en la interacción sean originales
        if (x_variables[i] in x_variables[:5]) and (x_variables[j] in x_variables[:5]):
            interaction_term = X[:, i] * X[:, j]
            interaction_term_name = x_variables[i] + 'x' + x_variables[j]
            X = np.column_stack((X, interaction_term))
            new_variable_names.append(interaction_term_name)

# Extender x_variables con las variables al cuadrado
x_variables.extend(new_variable_names)
```

```python
# Crear un DataFrame con las columnas adecuadas
column_names = ['M_squared', 'W_squared', 'H_squared', 'P_squared', 'S_squared',
                'MxW', 'MxH', 'MxP', 'WxH', 'WxP', 'HxP','WxS']

# Tomar solo las primeras 10 columnas de X que coinciden con las especificadas en column_names
X_df = pd.DataFrame(X[:, :12], columns=column_names)

# Concatenar las columnas de X_df a df
df = pd.concat([df, X_df], axis=1)
df.columns
```
